```python
def second_converter(x):
    span = 2
    split_words = x.split(" ")
    array = []
    total_seconds = 0

    for i in range(0, len(split_words), span):
        x = "".join(split_words[i:i+span])
        array.append(x)

    for i in array:
        if "MILLENNIA" in i:
            location = i.index("MILLENNIA")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 3.15576e+10
        if "CENTURY" in i:
            location = i.index("CENTURY")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 3.15576e+9
        if "DECADE" in i:
            location = i.index("DECADE")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 3.15576e+8
        if "YEARS" in i:
            location = i.index("YEARS")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 3.15576e+7
        if "MONTHS" in i:
            location = i.index("MONTHS")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 2.6298e+6
        if "DAYS" in i:
            location = i.index("DAYS")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 86400
        if "HOURS" in i:
            location = i.index("HOURS")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 3600      
        if "MINUTES" in i:
            location = i.index("MINUTES")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num * 60  
        if ("SECONDS" in i) & ("MILLISECONDS" not in i):
            location = i.index("SECONDS")
            slicer = slice(0, location)
            sliced_str = i[slicer]
            num = int(sliced_str)
            total_seconds += num
    return total_seconds

p15["Time_to_crack_in_seconds"] = p15["time"].apply(lambda x: second_converter(x))
```
